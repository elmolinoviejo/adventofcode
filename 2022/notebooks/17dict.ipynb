```python
'''
-------

..
'''

def get_next_rock(rn, yt):
    if rn==0:
        return [(2, yt),
                (3, yt),
                (4, yt),
                (5, yt)]
    if rn==1:
        return [(3, yt + 2),
                (2, yt + 1),
                (3, yt + 1),
                (4, yt + 1),
                (3, yt)]
    if rn==2:
        return [(4, yt + 2),
                (4, yt + 1),
                (2, yt),
                (3, yt),
                (4, yt)]
    if rn==3:
        return [(2, yt + 3),
                (2, yt + 2),
                (2, yt +1),
                (2, yt)]
    if rn==4:
        return [(2, yt + 1),
                (3, yt + 1),
                (2, yt),
                (3, yt)]

    
def move_down(r, rocks):
    new_r = [(x, y - 1) for x, y in r]
    if any([loc in rocks for loc in new_r]):
        # hits bottom
        for loc in r:
            rocks[loc] = '#'
        return 'next_rock', new_r, rocks
    else:
        return 'continue', new_r, rocks
    

def move_jet(r, rocks, j):
    if j=='>':
        dx = 1
    else:
        dx = -1
    new_r = [(x + dx, y) for x, y in r]
    if any([x in [-1, 7] for x, y in new_r]):
        # hit edge
        return r
    if any([loc in rocks for loc in new_r]):
        # hit rock
        return r
    else:
        return new_r
    
def draw_rocks(rocks):
    bottom = max([y for x, y in rocks])
    s = ''
    for y in reversed(range(0, bottom + 1)):
        row = ''
        for x in range(0, 7):
            if (x, y) in rocks:
                row +='#'
            else:
                row+='.'
        s +=f'{row}\n'
    return s


# def check_solid(rocks, last_tetris):
#     bottom = max([y for x, y in rocks])
#     for y_row in reversed(range(last_tetris + 2, bottom + 1)):
#         row = [(x, y_row) for x in range(0, 7)]
#         if all([loc in rocks for loc in row]):
#             # TETRIS
#             last_tetris = y_row
# #             print("--TETRIS--")
#             top_part = set()
#             for x in range(0, 7):
#                 for y in range(y_row, bottom + 1):
#                     top_part.add((x, y))
#             return rocks & top_part, last_tetris, True
#     return rocks, last_tetris, False

def trim_rocks(rocks, bottom, height=100):
#     bottom = max([y for x, y in rocks])
    new_rocks = {(loc[0], loc[1] - height): '#' for loc in rocks if loc[1] > bottom - height }
#     return 
#     for loc in rocks:
#         if bottom - 3000 < loc[1] < bottom + 1:
#             new_rocks[loc] = '#'
    return new_rocks, bottom - height
#     for y in range(bottom - 3000, bottom + 1):
        
```

```python
# input
problemnumber = '17'
fn = f'../{problemnumber}/{problemnumber}.txt'
# fn = f'../{problemnumber}/{problemnumber}t.txt'


with open(fn) as fh:
    jets = fh.read()
# print(len(jets))
n_repeat = len(jets)
jets = cycle(jets)




rocks = {(x, -1): '#' for x in range(7)}
rock_nos = cycle(range(5))
bottom = max([y for x, y in rocks])
yt = bottom + 4
rn = next(rock_nos)
r = get_next_rock(rn, yt)
n_stopped = 0

total_trimmed = 0 
# last_tetris = -1
# njets = 0
# nrocks = 1

while True:
    j = next(jets)
#     njets +=1
    # move jet
    r = move_jet(r, rocks, j)
    # move down
    cmd, r, rocks = move_down(r, rocks)
    if cmd == 'next_rock':
        bottom = max([y for x, y in rocks])
        yt = bottom + 4
        rn = next(rock_nos)
        r = get_next_rock(rn, yt)
#         nrocks += 1
        n_stopped += 1
    
        if n_stopped % 100 == 0:
            rocks, trimmed = trim_rocks(rocks, bottom, 100)
            total_trimmed += trimmed
#         print(len(rocks))

#         if n_stopped == 1000000000000:
        if n_stopped == 10002022:
            break
```
