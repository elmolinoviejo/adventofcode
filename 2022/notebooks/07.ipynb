```python

def load_input(fn):
    with open(fn) as fh:
        cmds = fh.read().split('\n')

    temp_ls = []
    fin_cmds = []
    for c in cmds:
        if '$ cd' in c:
            if len(temp_ls) > 0:
                fin_cmds.append(temp_ls)
            fin_cmds.append(c)
            temp_ls = []
        elif '$ ls' in c:
            continue
        else:
            temp_ls.append(c)
    
    if len(temp_ls) > 0:
        fin_cmds.append(temp_ls)

    cmds = fin_cmds
    current_loc = tuple('/') # dir
    G = [current_loc]
    sizes = {current_loc: 0}
    
    return cmds, G, current_loc, sizes

def run_cmd(cmd, G, current_loc, sizes):
    if '$ cd ' in cmd:
#         print(cmd)
        move_to = cmd[5:]
        if '..' in move_to:
            current_loc = current_loc[0:len(current_loc) - 1]
            return G, current_loc, sizes
        else:
            current_loc = tuple(list(current_loc) + [move_to])
            if current_loc not in sizes:
                sizes[current_loc] = 0
            return G, current_loc, sizes
    elif type(cmd) is list:
#         print(cmd)
        for f in cmd:
            size, name = f.split()
            if size=='dir':
                continue
            else:
                for l in range(1, len(current_loc) + 1):
                    sizes[tuple(list(current_loc)[:l])] += int(size)
#                 for d in current_loc:
#                     sizes[d] += int(size)                
            
        return G, current_loc, sizes
        
    
    return None
```

```python
# input
problemnumber = '07'
fn = f'../{problemnumber}/{problemnumber}.txt'

# fn = f'../{problemnumber}/{problemnumber}t.txt'
```

```python
cmds, G, current_loc, sizes = load_input(fn)
for cmd in cmds[1:]:
    G, current_loc, sizes = run_cmd(cmd, G, current_loc, sizes)
    print(current_loc)
```
