```python
from collections import Counter, deque
from itertools import product
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
day = '09'
fn = f'../{day}/{day}.txt'
# fn = f'../{day}/{day}_test.txt'

lmap = []
with open(fn) as fh:
    for l in fh.readlines():
        lmap.append(l.rstrip())
        

hs = dict()
for y, line in enumerate(lmap):
    for x, h in enumerate(line):
        hs[(int(x), int(y))] = int(h)
```

```python
def risk_level(loc, hs):
    return hs[loc] + 1


def get_adj(loc, hs):
    x, y = loc
    adj = []
    for a in [(x + 1, y),
                (x - 1, y),
                (x, y + 1),
                (x, y -1)]:
        if a in hs:
            adj.append(a)
    return adj


def check_low(loc, hs, exclude=[]):
    x, y = loc
    for adj in [(x + 1, y),
                (x - 1, y),
                (x, y + 1),
                (x, y -1)]:
        if (adj in hs) and (adj not in exclude):
            if hs[adj] <= hs[loc]:
                return False
    return True


def get_basin(loc, hs):
    tocheck = deque([loc])
    basin = []
    exclude = []
    while len(tocheck) > 0:
        print(tocheck, exclude)
        loc = tocheck.popleft()
        if hs[loc] == 9:
            continue
        exclude.append(loc)
        print(loc, hs[loc], check_low(loc, hs, exclude))
        if check_low(loc, hs, exclude):
            basin.append(loc)
            for a in get_adj(loc, hs):
                if (a not in exclude) and (a not in tocheck):
                    tocheck.append(a)
#             tocheck.extend([x for x in get_adj(loc, hs) if x not in exclude])
    return basin


def get_basin2(loc, hs):
    tocheck = deque([loc])
    basin = []
    exclude = []
    while len(tocheck) > 0:
#         print(tocheck, exclude)
        loc = tocheck.popleft()
        if hs[loc] == 9:
            continue
        elif hs[loc] < 9:
            basin.append(loc)
            exclude.append(loc)
            for a in get_adj(loc, hs):
                if (a not in exclude) and (a not in tocheck):
                    tocheck.append(a)
#         print(loc, hs[loc], check_low(loc, hs, exclude))
#         if check_low(loc, hs, exclude):
#             basin.append(loc)
#             for a in get_adj(loc, hs):
#                 if (a not in exclude) and (a not in tocheck):
#                     tocheck.append(a)
#             tocheck.extend([x for x in get_adj(loc, hs) if x not in exclude])
    return basin
        
#     x, y = loc
#     basin = [loc]
#     exlude = [loc]
#     check = []
#     for adj in [(x + 1, y),
#             (x - 1, y),
#             (x, y + 1),
#             (x, y -1)]:
#         if adj in hs:
#             check.append(adj)
    
    
```
