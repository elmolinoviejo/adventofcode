```python
from collections import Counter, deque
from itertools import product
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def load_input(fn):
    hs = {}
    with open(fn) as fh:
        lines = fh.read().split('\n')

    for y, line in enumerate(lines):
        for x, h in enumerate(line):
            hs[(int(x), int(y))] = int(h)
    return hs

def get_adj(loc, hs):
    x, y = loc
    adj = []
    for a in [(x + 1, y),
                (x - 1, y),
                (x, y + 1),
                (x, y -1),
                (x + 1, y + 1),
                (x - 1, y - 1),
                (x + 1, y - 1),
                (x - 1, y + 1)]:
        if a in hs:
            adj.append(a)
    return adj


def step(hs):
    n_flash = 0
    for loc, v in hs.items():
        hs[loc] += 1
    
    flashed = set()
    n_to_flash = len([v for loc, v in hs.items() if (v > 9) and (loc not in flashed)])
    while  n_to_flash > 0:
        for loc, v in hs.items():
            if v > 9:
                if loc in flashed:
                    continue
                n_flash += 1
                flashed.add(loc)
                hs[loc] = 0
                for adj in get_adj(loc, hs):
                    if adj in flashed:
                        continue
                    hs[adj] += 1
            n_to_flash = len([v for loc, v in hs.items() if (v > 9) and (loc not in flashed)])
    return hs, n_flash
            
```

```python
day = '11'
fn = f'../{day}/{day}.txt'
# fn = f'../{day}/{day}_test.txt'

hs = load_input(fn)

ans = 0
for _ in range(100):
    hs, nf = step(hs)
    ans += nf
```
