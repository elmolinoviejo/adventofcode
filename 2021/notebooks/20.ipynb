```python
from collections import Counter, deque, defaultdict
from itertools import product, count, combinations
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def load_input(fn):
    img_vals = set()
    with open(fn) as fh:
        enh, img = fh.read().split('\n\n')
        for y, line in enumerate(img.split('\n')):
            for x, v in enumerate(line):
                if v == '#':
#                     print(x,y)
                    img_vals.add(tuple([x, y]))

    return enh, img_vals

def get_neighbor_int_all(loc, img, allouton=False, verb=False):
    minx, maxx, miny, maxy = minmax(img)
    
    b = ''
    for dy, dx in product((-1, 0, 1), (-1, 0, 1)):
        x, y = loc
        n = tuple([x+dx, y+dy])
        
        outside = (x < minx) or (x > maxx) or (y < miny) or (y > maxy)
        if allouton and outside:
            b += '1'
        elif n in img:
            b += '1'
        else:
            b += '0'
    if verb:
        print(b)
    return int(b, 2)


def get_neighbor_int_all(loc, img, verb=False):
#     minx, maxx, miny, maxy = minmax(img)
    
    b = ''
    for dy, dx in product((-1, 0, 1), (-1, 0, 1)):
        x, y = loc
        n = tuple([x+dx, y+dy])
        
#         outside = (x < minx) or (x > maxx) or (y < miny) or (y > maxy)
#         if allouton and outside:
#             b += '1'
        if n in img:
            b += '1'
        else:
            b += '0'
    if verb:
        print(b)
    return int(b, 2)


def minmax(img):
    xs = [x for x, y in img]
    ys = [y for x, y in img]
    
    return min(xs), max(xs), min(ys), max(ys)


def print_img(img):
    minx, maxx, miny, maxy = minmax(img)
    out_str = ''
    for y in range(miny, maxy + 1):
        for x in range(minx, maxx + 1):
            if (x, y) in img:
                out_str += '#'
            else:
                out_str += '.'
        out_str += '\n'
    print(out_str)
```

```python
day = '20'
fn = f'../{day}/{day}.txt'
# fn = f'../{day}/{day}_test.txt'

enh, img = load_input(fn)


enh_s = set()
for ind, e in enumerate(enh):
    if e == '#':
        enh_s.add(ind)

print_img(img)
for ind in range(2): 
    
    minx, maxx, miny, maxy = minmax(img)
        
    if ind % 2 == 1:
        allouton = True
        for x, y in product(
                range(minx - 2, minx),
                range(miny -2, maxy + 3)):
            img.add((x, y))
        for x, y in product(
                range(minx - 2, maxx + 3),
                range(miny -2, miny)):
            img.add((x, y))
        for x, y in product(
                range(maxx+1, maxx + 3),
                range(miny -2, maxy + 3)):
            img.add((x, y))
        for x, y in product(
                range(minx - 2, maxx + 3),
                range(maxy + 1, maxy + 3)):
            img.add((x, y))
        print_img(img)
    
    
    
    new_img = set()
    for x, y in product(
            range(minx - 1, maxx + 2),
            range(miny - 1, maxx + 2)):
        
        
        val = get_neighbor_int((x, y), img)
        if val in enh_s:
            new_img.add(tuple([x, y]))
#     print(len(new_img))
    img = new_img
    print(ind)
        
len(img)
```
