```python
fn = '../02/02.txt'

inputs = []
with open(fn) as fh:
    for line in fh.readlines():
        inputs.append(tuple(line.rstrip().split(' ')))

inputs
```

```python
start = (0, 0)
loc = start
for d, n in inputs:
    n = int(n)
    if d=='up':
        loc = (loc[0], loc[1] - n)
    elif d=='down':
        loc = (loc[0], loc[1] + n)
    elif d=='forward':
        loc = (loc[0] + n, loc[1])

        
    
```

```python
loc[0] * loc[1]
```

```python
start = (0, 0)
loc = start
aim = 0


for d, n in inputs:
    n = int(n)
    if d=='up':
        aim -= n
#         loc = (loc[0], loc[1] - n)
    elif d=='down':
        aim += n
#         loc = (loc[0], loc[1] + n)
    elif d=='forward':
        loc = (loc[0] + n, loc[1] + (aim * n))

        
    
```
