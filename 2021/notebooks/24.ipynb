```python
def load_input(fn):
    inst = []
    with open(fn) as fh:
        for l in fh.readlines():
            cmd, *vals = l.rstrip().split()
            inst.append((cmd, vals))
    return inst


def load_input_blocks(fn):
    blocks = {}
    blockn = -1
    inst = []
    with open(fn) as fh:
#         b_strs = fh.read().split('inp')
#         print(b_strs)
        for l in fh.readlines():
            cmd, *vals = l.rstrip().split()
            if cmd == 'inp':
                if blockn >=0:
                    blocks[blockn] = inst
                inst = []
                blockn += 1
            else:
                inst.append((cmd, vals))
    blocks[blockn] = inst
    return blocks


def load_input_blocks(fn):
    blocks = {}
    blockn = -1
    inst = []
    with open(fn) as fh:
#         b_strs = fh.read().split('inp')
#         print(b_strs)
        for l in fh.readlines():
            cmd, *vals = l.rstrip().split()
            if cmd == 'inp':
                if blockn >=0:
                    blocks[blockn] = inst
                inst = []
                blockn += 1
            else:
                inst.append((cmd, vals))
    blocks[blockn] = inst
    return blocks


        

def inst_to_str(inst):
    cmd, vals = inst
    if cmd == 'add':
        return f'{vals[0]} = {vals[0]} + {vals[1]}'
    elif cmd == 'mul':
        return f'{vals[0]} = {vals[0]} * {vals[1]}'
    elif cmd == 'eql':
        return f'{vals[0]} = f({vals[0]}, {vals[1]})'
    elif cmd == 'div':
        return f'{vals[0]} = {vals[0]} / {vals[1]}'
    elif cmd == 'mod':
        return f'{vals[0]} = {vals[0]} % {vals[1]}'
    elif cmd == 'inp':
        return f'w = inp{vals[0][1:]}'
    
    
def run_inst(inst, vs):
    cmd, vals = inst
    val_ints = []
    for val in vals:
        if val in 'wxyz':
            val_ints.append(vs[val])
        else:
            val_ints.append(int(val))
            
    if cmd == 'add':
        vs[vals[0]] = sum(val_ints)
    elif cmd == 'mul':
        vs[vals[0]] = val_ints[0] * val_ints[1]
    elif cmd == 'eql':
        if val_ints[0] == val_ints[1]:
            ans = 1
        else:
            ans = 0
        vs[vals[0]] = ans
    elif cmd == 'div':
        vs[vals[0]] = int(val_ints[0] / val_ints[1])
    elif cmd == 'mod':
        vs[vals[0]] = val_ints[0] % val_ints[1]
    
    return vs


def run_block(block, vs):
    for inst in block:
        vs = run_inst(inst, vs)
    return vs


def test_zs(z, block):
    wzs = {}
    for w in range(1, 10):
        vs = {'w': 0, 'x': 0, 'y': 0, 'z': z}
        vs['w'] = w
        vs = run_block(block, vs)
        wzs[w] = vs['z']
    return wzs
```
