```python
import numpy as np

day = '05'
fn = f'../{day}/{day}.txt'
# fn = f'../{day}/{day}_test.txt'


with open(fn) as fh:
    lines = fh.read().split('\n')

edge_locs = []    
for l in lines:
    edges = l.split(' -> ')
    e1 = [int(x) for x in edges[0].split(',')]
    e2 = [int(x) for x in edges[1].split(',')]
    edge_locs.append([tuple(e1), tuple(e2)])
```

```python
# check if diag
def check_diag(edge):
    (x1, y1), (x2, y2) = edge
    if (x1 != x2) & (y1 != y2):
        return True
    return False

def get_edge_locs(edge):
    (x1, y1), (x2, y2) = edge
    xa = min(x1, x2)
    xb = max(x1, x2)
    ya = min(y1, y2)
    yb = max(y1, y2)
    locs = []
    if xa == xb:
        for y in range(ya, yb+1):
            locs.append((xa, y))
    elif ya == yb:
        for x in range(xa, xb+1):
            locs.append((x, ya))
    else:
        xstep, ystep = 1, 1
        if x1 > x2:
            xa = x1
            xb = x2 - 1
            xstep = -1
        elif x1 < x2:
            xa = x1
            xb = x2 + 1
            
        if y1 > y2:
            ya = y1
            yb = y2 - 1
            ystep = -1
        elif y1 < y2:
            ya = y1
            yb = y2 + 1
#         print(xa, xb, step, ya, yb, step)
        for x, y in zip(np.arange(xa, xb, xstep), np.arange(ya, yb, ystep)):
            locs.append((x, y))
    return locs
```

```python
loc_counts = dict()
# maxx, maxy = 0, 0
for el in edge_locs:
    if check_diag(el):
        continue
    (x1, y1), (x2, y2) = el
    locs = get_edge_locs(el)
    for l in locs:
        if l not in loc_counts:
            loc_counts[l] = 1
        else:
            loc_counts[l] += 1
```

```python
count = 0
for l, c in loc_counts.items():
    if c > 1:
        count+=1
count
```
